```python
import pandas as pd
import pm4py
```

```python
df=pd.read_csv(r"C:\Users\sneju\iiit python\file.csv\ecommerce_order_log_raw.csv")
print("===Step1:Raw data loaded ===")
print("shape:",df.shape)
print("Unique Orders:",df['OrderID'].nunique())
print(df.head(10))
print()
```

```python
print(df.columns)
```

```python
event_log = pm4py.format_dataframe(
    df, 
    case_id='OrderID', 
    activity_key='Activity', 
    timestamp_key='Timestamp'
)
print("===Step2: Formatted as a PM4Py event log ===")
print("Columns:",event_log.columns.tolist())
print()
```

```python
print("end===Step3: Data Quality Check ===")
print("Missing Values per columns:")
print(event_log.isnull().sum())
print("Duplicates:",event_log.duplicated().sum())

already_sorted=event_log.equals(
event_log.sort_values(["case:concept:name","time:timestamp"]).reset_index(drop=True)
)
print("Already sorted by case and timestamp:",already_sorted )
print()
```

```python
event_log = event_log.drop_duplicates()
event_log["Resource"]=event_log["Resource"].fillna("Unknown")
event_log=event_log.sort_values(
    ["case:concept:name","time:timestamp"]
    ).reset_index(drop=True)
print("===Step4:Cleaned ===")
print("Shape after cleaning:",event_log.shape)
print("Remaining missing values:",event_log.isnull().sum().sum())
print("Remaining duplicates:",event_log.duplicated().sum())
print()
```
