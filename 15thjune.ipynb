```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import(
    mean_absolute_error,#Average prediction mistake made by the model...actual-predicted
    mean_squared_error,#Average squared mistake made by the model.....To punish large mistakes more heavily...10² = 100
    r2_score#How much of the data pattern the model understands.
)

#Lower MAE,mse = Better Model
#r2 higher= better value
#range
#1.0  = Perfect
#0.9  = Excellent
#0.8  = Very Good
#0.7  = Good
#0.5  = Average
#0    = No Learning
#<0   = Very Bad Model
```

```python

#loading the data
#to study the data for any information
df=pd.read_csv("file.csv\Regression_Dataset_1000.csv")
df
```

```python
df.head(5)
```

```python
#features and target
x=df[
        "Study_Hours",
        "Attendance",
        "Assignments_Completed"
]
y=df["Exam_Score"]
```

```python
print (x)
```

```python
print (y)
```

```python
#train and split
x_train,x_test,y_train,y_test=train_test_split(# ek set for train and ek set for testing
    x,
    y,
    test_size=0.20, #for splitting in 20 /80 ...if we want 70/30 theh size will be 0.30
    random_state=42 #after prediction it compares with the actual data
)#computational value where we randomly select the data for splitting without any vedbhav diffferentiation
```

```python
#model training
model=LinearRegression()
model.fit(
    x_train,
    y_train
)
```

```python
#prediction
y_pred=model.predict(x_test)
```

```python
#evaluation
mae=mean_absolute_error(
    y_test,
    y_pred
)
mse=mean_squared_error(
    y_test,
    y_pred
)

rmse=np.sqrt(mse)#The value becomes too large and is no longer in the original unit,So we take the square root:Now the error is back in the original unit.

r2=r2_score(
    y_test,
    y_pred
)

print("\nMAE=",mae)
print("MSE=",mse)
print("RMSE=",rmse)
print("R2 Score=",r2)
```

```python
#actual vs predicted
results = pd.DataFrame({#used to create table
    "Actual": y_test,
    "Predicted": y_pred
})

print(results.head())
```
