```python
# Fill missing values in numerical columns using Median
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
#. Fill missing values in categorical columns using Mode
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df = df.drop_duplicates()
```

```python
df["Installs"] = (
    df["Installs"]
    .str.replace(",", "")
    .str.replace("+", "", regex=False)
)

df["Installs"] = df["Installs"].astype(int)
```

```python
df["Price"] = (
    df["Price"]
    .str.replace("$", "", regex=False)
)

df["Price"] = df["Price"].astype(float)
```

```python
df["Reviews"] = pd.to_numeric(df["Reviews"])
```

```python
def clean_size(x):
    if x == "Varies with device":
        return np.nan

    if "M" in x:
        return float(x.replace("M","")) * 1024

    if "k" in x:
        return float(x.replace("k",""))

    return np.nan

df["Size"] = df["Size"].apply(clean_size)
```

```python
df["Size"] = df["Size"].fillna(df["Size"].median())
```

```python
df = df[
    (df["Rating"] >= 0) &
    (df["Rating"] <= 5)
]
```

```python
Q1 = df["Reviews"].quantile(0.25)
Q3 = df["Reviews"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[
    (df["Reviews"] >= lower) &
    (df["Reviews"] <= upper)
]
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])

df["Update_Year"] = df["Last Updated"].dt.year
```
